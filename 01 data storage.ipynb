
# database

a structured collection of data that is stored and managed electronically. It allows for - efficient storage, retrieval, modification, and deletion of data

# data warehouse

a centralized repository of structured data collected from various data source to support analytical queries

# data mart

a subset of a data warehouse, specifically designed to serve the needs of a particular department or team

# data lake

 is a centralized repository of raw, unstructured, semi-structured, and structured data in its native format and unprocessed state

data lakehouse

is a data architecture that combines the best of data lakes and data warehouses. It provides a unified platform for storing, processing, and analyzing structured and unstructured data at scale, using open standards and cloud-native technologies

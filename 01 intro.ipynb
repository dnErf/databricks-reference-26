### data
a unit or collection of observation that can be used as a basis for reasoning

### data type
a single unit of data that tells a compiler or interpreter how data is intended to be used

##### numeric data type
involving mathematical numbers
  - integer
  - float

##### text data types
contains readable and non-readable letters
  - character
  - string

##### composite
contains cells of data that can be accessed via an index or a key
  - array
  - hash

##### binary data type
represented by a bit

##### boolean
data type that is either True or False

##### enum
group of constant variables

### schema
a blueprint for which describes the structure of data of a database

### schemaless
flexible dynamic data models without predefined structure

### data documents
defines the collective form in which data exists, an abstract groupings of data

- databases
- datastores
- data warehouses
- notebooks

### data sets
unstructured logical grouping of data that genrally are closely related, share same data structure

### query
a request to perform operations to the database

### querying
act of performing a query

### query language
designed to interact with information systems such as database to retrieve, manipulate and manage data

### batch processing
processing large volumes of data over a fixed period before processing it all at once

### strem processing
process the data as it arrives
